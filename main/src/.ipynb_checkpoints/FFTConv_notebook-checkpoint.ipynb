```python
file_path = '../resources/genres/hiphop/hiphop.00000.wav'
x , sampling_rate = librosa.load(file_path)
```

```python
plt.figure(figsize=(20, 6))
librosa.display.waveplot(x, sr=sampling_rate)
```
